```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy.ndimage import gaussian_filter
```

# a) Thêm nhiễu Gaussian vào ảnh "coins.tif"

```python
image = io.imread("coins.tif")  
noisy_image = util.random_noise(image, mode='gaussian', var=0.01) 
```

# b) Thực hiện lọc Gaussian cho ảnh nhiễu

```python
image
```

```python
sigma = 1  
filtered_image = gaussian_filter(noisy_image, sigma=sigma)
```

# c) Vẽ histogram và tìm điểm ngưỡng

```python
def histogram(img):
#create list of values 0-255
    pixels=[]
    counts=[0]*256
    for x in range(256):
        pixels.append(x)
    for h in range(len(img)):
        for w in range(len(img[0])):
            counts[img[h, w]] +=1
    plt.figure()
    plt.bar(pixels,counts)
    plt.show()
```

```python
histogram(image)
```

```python
threshold=200
binary_image = image > threshold 
```

# d) Sử dụng điểm ngưỡng để phân tách vật thể và nền

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Ảnh Gốc")

axes[1].imshow(noisy_image, cmap='gray')
axes[1].set_title("Ảnh Thêm Nhiễu Gaussian")

axes[2].imshow(filtered_image, cmap='gray')
axes[2].set_title("Ảnh Sau Lọc Gaussian")

axes[3].imshow(binary_image, cmap='gray')
axes[3].set_title("Ảnh Phân Tách Vật Thể và Nền")

for ax in axes:
    ax.axis("off")

plt.tight_layout()
plt.show()
```
